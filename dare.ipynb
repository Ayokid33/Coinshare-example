```python
import os, glob
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt

path = "C:/Users/ayoki/OneDrive/Desktop/20230201damlbmp_zone_csv"

all_files = glob.glob(os.path.join(path, "*.csv"))

all_df = []
for f in all_files:
    df = pd.read_csv(f, sep=',')
    df['file'] = f.split('/')[-1]
    all_df.append(df)
    
merged_df = pd.concat(all_df, ignore_index=True, sort=True)
merged_df.pop('file')
#removed the useless clutter column called file.

merged_df.to_csv("merged22.csv")
#saving merged file as name above
```
